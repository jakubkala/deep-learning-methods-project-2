```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict

def get_image(img: np.ndarray):
    """
    input: img should be a row of cifar-100 dataset.
    returns: numpy 3D array, to be used with plt.imshow() function.
    """
    return np.reshape(img, (32, 32, 3), order='F').transpose(1, 0, 2)
```

```python
raw_data = unpickle('../data/data_batch_1')
X_train = raw_data[b'data']
y_train = np.array(raw_data[b'labels'])
```

```python
plt.imshow(get_image(X_train[1]))
```

```python
def change_contrast(image, value):
    falue = float(value)
    return np.clip(128 + image * value - 128 * value, 0, 255).astype(np.uint8)

def change_brightness(image, value):
    falue = float(value)
    return np.clip(image * value, 0, 255).astype(np.uint8)
```

```python
plt.imshow(change_contrast(get_image(X_train[6]), 1.4))
```

```python
plt.imshow(change_contrast(get_image(X_train[6]), 0.6))
```

```python
plt.imshow(change_brightness(get_image(X_train[1]), 1.3))
```

```python
plt.imshow(change_brightness(get_image(X_train[1]), 0.7))
```
